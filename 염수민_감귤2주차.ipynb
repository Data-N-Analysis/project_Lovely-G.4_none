```python
new_columns=[]
for col in train.columns:
    if '새순' in col:
        new_columns.append(col)
```

```python
new_columns2=[]
for col in train.columns:
    if '엽록소' in col:
        new_columns.append(col)
```

```python
train['새순mean']= train.iloc[:, 6:95].mean(axis=1)
train["새순std"]= train.iloc[:, 6:95].std(axis=1)
train["새순diff"]= train["2022-11-28 새순"] - train["2022-09-01 새순"]
train["새순min"]= train.iloc[:, 6:95].min(axis=1)
train["새순max"]= train.iloc[:, 6:95].max(axis=1)
```

```python
train["엽록소mean"]= train.iloc[:, 96:184].mean(axis=1)
train["엽록소std"]= train.iloc[:, 96:184].std(axis=1)
train["엽록소diff"]= train["2022-11-28 엽록소"] - train["2022-09-01 엽록소"]
train["엽록소min"]= train.iloc[:, 96:184].min(axis=1)
train["엽록소max"]= train.iloc[:, 96:184].max(axis=1)
```

```python
train=train.drop(new_columns,axis=1)
```

```python
train=train.drop(new_columns2,axis=1)
```

```python
# 여러번 반복할 수 있도록 함수 따로 원본 저장
original_train = train
original_test = test

# 첫 번째 열(ID) 제외
train = train.drop(columns = ['ID'])
```

```python
# 분산팽창계수(VIF) 확인해보기!

from statsmodels.stats.outliers_influence import variance_inflation_factor

train['Intercept'] = 1
# 상수항을 추가함으로써 모델이 항상 절편을 고려하게 되어, 각 독립 변수가 선형 종속이더라도 공분산 행렬이 올바르게 형성되어 다중공선성을 평가할 수 있게 됨.

# VIF 계산
vif_data = pd.DataFrame()
vif_data["Variable"] = train.columns
vif_data["VIF"] = [variance_inflation_factor(train.values, i) for i in range(train.shape[1])]

# 결과 출력
print(vif_data)
```

```python
exclude_v = ['수관폭1(min)']

# 선택된 변수들을 제외한 나머지 변수들을 가진 데이터프레임 생성
train = train.drop(exclude_v, axis=1)

# VIF 계산
vif_data = pd.DataFrame()
vif_data["Variable"] = train.columns
vif_data["VIF"] = [variance_inflation_factor(train.values, i) for i in range(train.shape[1])]

# 결과 출력
print(vif_data)
```

```python
exclude_v = ['수관폭2(max)']

# 선택된 변수들을 제외한 나머지 변수들을 가진 데이터프레임 생성
train = train.drop(exclude_v, axis=1)

# VIF 계산
vif_data = pd.DataFrame()
vif_data["Variable"] = train.columns
vif_data["VIF"] = [variance_inflation_factor(train.values, i) for i in range(train.shape[1])]

# 결과 출력
print(vif_data)
```

```python
exclude_v = ['엽록소max']

# 선택된 변수들을 제외한 나머지 변수들을 가진 데이터프레임 생성
train = train.drop(exclude_v, axis=1)

# VIF 계산
vif_data = pd.DataFrame()
vif_data["Variable"] = train.columns
vif_data["VIF"] = [variance_inflation_factor(train.values, i) for i in range(train.shape[1])]

# 결과 출력
print(vif_data)
```
